```python
iris2 = iris[[
    'sepal.length',
    'sepal.width',
    'petal.length',
    'petal.width'
]]

#v =np.random.random((150, ))
def create_reducer(v) :
    def reducer(x):
        return x @ v
    return reducer

vetor_1 = np.random.random((150, )) * 10 # eixo de linha
vetor_2 = np.random.random((4 , )) * 10   # eixo de coluna

iris2.apply(create_reducer(vetor_1), axis=0) # eixo de linha
iris2.apply(create_reducer(vetor_2), axis=1) # eixo de coluna
```

```python
from sklearn.model_selection import train_test_split
from sklearn.neural_network import MLPClassifier
```

```python
features = [
    'sepal.length',
    'sepal.width',
    'petal.length',
    'petal.width',
] 

target = 'variety'

X_train, X_test, y_train, y_test = train_test_split(
                                                   iris[features],
                                                   iris[target],
                                                   test_size=0.33,
                                                   random_state=np.random.randint(0,100))
```

```python
clf = MLPClassifier() 
```

```python

from sklearn.metrics import accuracy_score
clf.fit(X_train, y_train)
predctions = clf.predict(X_test)
accuracy_score(predctions, y_test)
```
